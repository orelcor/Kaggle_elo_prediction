```python
train_df = pd.read_csv('../input/train.csv', header = 0)
train_df.head()
```

# Univariate analysis

```python
sns.distplot(train_df.target)
```

```python
train_df = train_df[train_df.target > -30] # replace outliers in target
```

```python
train_df.target.describe() # like normal with mean = 0 and std = const
```

```python
sns.countplot(train_df.feature_1)
```

```python
sns.countplot(train_df.feature_2)
```

```python
sns.countplot(train_df.feature_3)
```

# Multivariate analysis

```python
gb_feature_1 = train_df.groupby(by = 'feature_1')['target']
gb_feature_1.mean().plot()
```

```python
gb_feature_1.median().plot()
```

```python
gb_feature_2 = train_df.groupby(by = 'feature_2')['target']
gb_feature_2.mean().plot()
```

```python
gb_feature_2.median().plot()
```

```python
gb_feature_3 = train_df.groupby(by = 'feature_3')['target']
gb_feature_3.mean().plot()
```

```python
gb_feature_3.median().plot()
```
